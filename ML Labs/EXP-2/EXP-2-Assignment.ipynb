### The dataset contains the record of the profit earned by a food truck in cities with population mentioned in the first column. Construct a linear regression model that can predict the profit if the population of a city is known.

Import Libraries Pandas for Dataframe, Mathplotlib for graph, sklearn for regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, r2_score
```

1) Read Data from the CSV

```python
df = pd.read_csv(r'profits.csv', names=['population', 'profit'])
```

2) Extract features and target for linear regression

- `x = df[['population']]` returns a **2D DataFrame**, which is required by scikit-learn models as input (`shape: [n_samples, n_features]`).
- `y = df['profit']` returns a **1D Series**, which is suitable as the target vector (`shape: [n_samples]`).

⚠️ Using `df['population']` instead of `df[['population']]` would return a 1D Series, which may cause errors during model fitting.

```python
x = df[['population']]
y = df['profit']
```

3) Check the data for any formatting error

```python
df.head()
```

4) Initialize Linear Regression Model from SKLearn

```python
regression_model = LinearRegression()
regression_model.fit(x,y)
```

5) Store Model Prediction

```python
y_predicted = regression_model.predict(x)
```

6) Initialize Root Mean Squared error

```python
rmse = root_mean_squared_error(y,y_predicted)
r2 = r2_score(y,y_predicted)
```

7) Print Slope, Intercept, RMSE, R2 Score

```python
print('Slope:',regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error:', rmse)
print('R2 score:',r2)
```

8) Plot the graph using matplotlib

```python
plt.scatter(x,y)
plt.plot(x,y_predicted, color='r')
plt.xlabel('Population')
plt.ylabel('Profit')

plt.show()
```
