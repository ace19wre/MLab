```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as exp

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r"Country_data.csv")
df.head()
```

```python
df_no_country = df.drop('country', axis=1)
```

```python
df_no_country.corr()
```

```python
scalar = MinMaxScaler()
```

```python
df.drop('country', axis=1)
```

```python
scaled_data = scalar.fit_transform(df_no_country)
```

```python
scaled_df = pd.DataFrame(data=scaled_data, columns=df_no_country.columns)
```

```python
scaled_df['country'] = df['country']
```

```python
scaled_df.head()
```

```python
data = scaled_df.drop('country', axis = 1)
```

```python
km = KMeans(n_clusters=5)
km = km.fit(data)
```

```python
ssd = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data)
    ssd.append(km.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```
